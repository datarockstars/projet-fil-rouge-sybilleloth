```python
# Supprimer les lignes où les colonnes 'longitude' et 'latitude' sont vides
cities_df = cities_df.dropna(subset=['longitude', 'latitude'])

# Afficher les premières lignes du DataFrame pour vérifier
print(cities_df.head())
```

```python
# Supprimer les colonnes redondantes
cities_df = cities_df.loc[:, ~cities_df.columns.duplicated()]

# Afficher les colonnes restantes pour vérifier
print(cities_df.columns)
```

```python
# Supprimer les lignes identiques
cities_df = cities_df.drop_duplicates()

# Afficher les premières lignes du DataFrame pour vérifier
print(cities_df.head())
cities_df.shape
```

```python
# Supprimer les lignes où 'longitude' et 'latitude' sont identiques à d'autres, en gardant uniquement la première occurrence
cities_df = cities_df.drop_duplicates(subset=['longitude', 'latitude'], keep='first')

# Afficher les premières lignes du DataFrame pour vérifier
print(cities_df.head())
```

```python
cities_df.shape
```

```python
# Convert column names to lowercase and replace hyphens with spaces
esr_df['Académie modifiée'] = esr_df['Académie'].str.lower().str.replace('-', ' ')
esr_df['Académie_modifiée'] = (
    esr_df['Académie']
    .str.lower()
    .str.replace('-', ' ')
    .str.replace('é', 'e')
    .str.replace('è', 'e')
)

replacement_mapping = {
    'lyon': 'lyon 01',
    'orleans tours': 'orleans',
    'paris': 'paris 01',
    'besançon': 'besancon',
    'nancy metz': 'nancy',
    'aix marseille': 'marseille 01',
    'créteil': 'creteil',
    'normandie': 'rouen',
    'la réunion': 'st denis',
    'guyane': 'cayenne',
    'guadeloupe': 'pointe a pitre',
    'corse': 'corte'
 }

# Remplacer les valeurs dans la colonne 'Académie'
esr_df['Académie modifiée'] = esr_df['Académie_modifiée'].replace(replacement_mapping)


# Merge DataFrames on city_code
merged_df = pd.merge(esr_df, cities_df, left_on='Académie modifiée', right_on='city_code', how='left')

# Select necessary columns
columns_to_keep = [
    'Région', 'Académie', 'Académie modifiée', 'Établissement', 'Établissement actuel', 
    'Type de diplôme', 'Domaine disciplinaire', 'Discipline', 'Secteur disciplinaire',
    'Libellé du diplôme', 'Nombre de poursuivants', 'Nombre de sortants',
    'Année(s) d\'obtention du diplôme prise(s) en compte', 'Flag', 
    'Taux d\'emploi salarié en France', 'Taux d\'insertion', 'Taux d\'emploi',
    'latitude', 'longitude', 'department_name', 'region_name', 'region_geojson_name'
]

final_df = merged_df[columns_to_keep]

# Supprimer la colonne 'Académie modifiée' du DataFrame
final_df = final_df.drop(columns=['Académie modifiée'])

# Afficher les premières lignes du DataFrame pour vérifier
print(final_df.head())

# Save to new CSV
output_path = "esr_nettoye_avec_cities.csv"
final_df.to_csv(output_path, index=False)
print(f"Le fichier de fusion a été créé : {output_path}")
```
